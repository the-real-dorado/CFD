```python
import CFD
```

```python
file = 'meshes/circle.obj'
domain = CFD.domain(file,0.001,1000,500)
```

```python
fluid = CFD.fluid(domain)
model = CFD.models.flow(domain,0.1,0)
CFD.simulation(fluid,model,dt=1e-5,rt=0.005)
```

```python
CFD.data.plot(fluid,['pressure'])
CFD.data.plot(fluid,['velocity','streamlines'])
```
